```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import numpy as np
import pandas as pd 
```

```python
app = dash.Dash()
```

```python
orders=pd.read_excel('Sample-Superstore.xls')
orders.head()
```

```python
# Wrapping dash core components into the html components
app.layout= html.Div([
    dcc.Graph(
        id='Sales vs Profit Data',
        figure={
            'data':[
                go.Scatter(
                    x=orders['Sales'],
                    y=orders['Profit'],
                    mode='markers'
                          )
                   ],
             'layout':go.Layout(
                 title='Sales vs Profit',
                 xaxis={'title':'Sales'},
                 yaxis={'title':'Profit'},
                 hovermode='closest'
                 
                               )
            
            
                }
    
              )
    
    
    
])
```

```python
if __name__ == '__main__':
    app.run_server(port=4051)
```
