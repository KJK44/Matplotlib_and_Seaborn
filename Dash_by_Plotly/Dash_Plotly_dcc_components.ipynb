```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import numpy as np
import pandas as pd
from datetime import datetime as dt
```

#### Components
     - Dropdown
     - Slider
     - Range Slider
     - Input Box

```python
app=dash.Dash()
```

```python
app.layout=html.Div([
    
    dcc.Markdown( '''
        #### Dash by Plotly - `Markdown`
        Dash supports [Markdown](https://plotly.com/dash/) \n
        It includes **Bold Text** : use double asterisks. \n
        Also *Italic Text* : use single asterisks. \n
       '#' : for headings. 
       '''
                ),
    
    html.Label("Dropdown :"),   # As we are using placeholder option, This is no longer needed
    html.Br(),
    html.Br(),
    dcc.Dropdown(
        id='Dropdown',
        options=[
            {'label':'San Fransisco', 'value':'SF'},
            {'label':'New York City','value':'NYC'},
            {'label': 'Newark','value':'NWK','disabled':True}          
                ],
         placeholder='Select a City'
         #disabled=True  # This disables the dropdown. It shows on the screen but is disabled.
         #value='NYC',   # This is default
         # multi=True   # This argument is for allowing multiple selection
         
        
                ), #dcc.Dropdown
    
    html.Br(),
    html.Br(),
    html.Label("Slider : "),
    dcc.Slider(
        min=1,
        max=10,
        value=5,
        step=0.5,
        marks={i : i for i in range(10)}
        
              ),
    html.Br(),
    html.Br(),
    html.Label("Range Slider :"),
    dcc.RangeSlider(
        min=1,
        max=10,
        step=0.5,
        value=[2,7],
        marks={i : i for i in range(10)}
                    ),
    
   
    html.Br(),
    html.Br(),               
    html.Label("Input Box :"),
    dcc.Input(
        placeholder='Input Your Name',
        type='Text',
        value=''
        
             ),
        
        
    html.Br(),
    html.Br(),
    html.Br(),
    
    html.Label("Text Area :"),    
    dcc.Textarea(
        placeholder='Input your feedback',
        #value='Placeholder for Text',
        style={'width':'50%'}
                ), #Textarea
    html.Br(),
    html.Button('Submit',id='submit-form'),
    
    html.Br(),
    html.Br(),
    html.Br(),
    html.Label("Check List :"),    
    dcc.Checklist(
        options=[
            {'label':'San Fransisco', 'value':'SF'},
            {'label':'New York City','value':'NYC'},
            {'label': 'Newark','value':'NWK'}          
                ],
        value=['SF','NYC']
                 ),
        
    
    html.Br(),
    html.Br(),
    html.Label("Radio Button :"),    
    dcc.RadioItems(
        options=[
            {'label':'San Fransisco', 'value':'SF'},
            {'label':'New York City','value':'NYC'},
            {'label': 'Newark','value':'NWK'}          
                ],
        value='SF'
                 ),
    
    html.Br(),
    html.Br(),
    html.Label("Date Picker Single : "),
    dcc.DatePickerSingle(
        id='date picker single',
        date=dt(2020,5,10)
                        ),
    
    html.Br(),
    html.Br(),
    html.Label("Date Picker Range : "),
    dcc.DatePickerRange(
        id='date picker range',
        start_date=dt(2020,5,10),
        end_date_placeholder_text='end date'
                        )
    
  
    
   
    
                
])
```

```python
if __name__ == '__main__':
    app.run_server(port=4050)
```
