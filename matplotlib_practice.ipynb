**Reading csv file using csv module - csv.DicReader()**

```python
import csv
```

```python
with open(dir_path+'data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

languages = []
popularity=[]

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
    
print(languages)
print("*****")
print(popularity)
print("*****")
# print(language_counter)
print(language_counter.most_common(15)[0])   # This gives us top 15
print(type(language_counter.most_common(15)[0])) 
#     row = next(csv_reader)
#     row
#     print(row,'\n',row['LanguagesWorkedWith'],'\n',)


languages.reverse()
popularity.reverse()
```

```python
plt.title("Most Popular Languages")
plt.ylabel("Programming Languages")
plt.xlabel("Number of People")
plt.barh(languages,popularity)
```
